Hồi quy logistic: thư viện scikit-learn
Sử dụng thư viện scikit-learn và matplotlib trong Python, bạn hãy thực hiện các yêu cầu sau:
Khởi tạo dữ liệu:
- Tạo một mảng NumPy X gồm các điểm dữ liệu 2D như sau: [[1, 2], [2, 3], [3, 1], [2, 5], [4, 3]].
- Tạo một mảng NumPy Y tương ứng với nhãn của các điểm dữ liệu, trong đó 1 là một lớp và 0 là lớp còn lại: [1, 1, 1, 0, 0].
Xây dựng và huấn luyện mô hình:
- Sử dụng LogisticRegression để xây dựng một mô hình hồi quy logistic.
- Huấn luyện mô hình với dữ liệu X và nhãn Y.
Dự đoán và trực quan hóa:
- Dự đoán nhãn cho một điểm dữ liệu mới X_pred = [1, 4].
- Trích xuất các hệ số w1, w2 và hằng số b từ mô hình đã huấn luyện.
Vẽ biểu đồ minh họa kết quả, bao gồm:
- Các điểm dữ liệu ban đầu, phân biệt bằng màu sắc dựa trên nhãn của - chúng.
- Điểm dữ liệu mới cần dự đoán.
- Vẽ đường phân tách của mô hình hồi quy logistic.

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Khởi tạo dữ liệu
X = np.array([[1, 2],
                [2, 3],
                [3, 1],
                [2, 5],
                [4, 3]])
Y = np.array([1, 1, 1, 0, 0])

# Xây dựng và huấn luyện mô hình Logistic Regression
model = LogisticRegression()
model.fit(X, Y)

# Dự đoán một điểm mới
X_pred = np.array([[1, 4]])
Y_pred = model.predict(X_pred)

# Lấy các hệ số và hằng số của mô hình
w1, w2 = model.coef_[0]
b = model.intercept_[0]

# --- Vẽ biểu đồ ---
plt.figure()

# (1) Vẽ các điểm gán nhãn 1
plt.scatter(
    X[Y == 1][:, 0], X[Y == 1][:, 1],
    color='blue', label='Các điểm gán nhãn 1'
)

# (2) Vẽ các điểm gán nhãn 0
plt.scatter(
    X[Y == 0][:, 0], X[Y == 0][:, 1],
    color='red', label='Các điểm gán nhãn 0'
)

# (3) Vẽ các điểm mới với nhãn dự báo
plt.scatter(
    X_pred[0, 0], X_pred[0, 1],
    color='purple', marker='x', s=100,
    label='Điểm với nhãn dự báo là: 1'
)

# (4) Xây dựng đường phân tách
a = -w1 / w2
b_ = -b / w2
X1_grid = np.linspace(0, 5, 100)
X2_grid = a * X1_grid + b_
plt.plot(
    X1_grid, X2_grid,
    color='green', label='Đường phân tách'
)

# Gán nhãn cho các thành phần, đặt tên trục và tiêu đề
plt.legend()
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

# Hiển thị biểu đồ
plt.show()
```
