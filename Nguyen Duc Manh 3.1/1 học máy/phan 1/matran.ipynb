```python
A = np.concatenate((np.ones(shape=(len(X), 1)), X), axis= 1)
A
```

```python
np.linalg.det(A.T @ A)
```

```python
U, S, Vh = np.linalg.svd(A, full_matrices=False)
```

```python
U
```

```python
S
```

```python
Vh 
```

```python
V = Vh.T
V
```

```python
S_inv = [1/s if s > 1e-6 else 0 for s in S ]
S_inv
```

```python
V @ np.diag(S_inv) @ U.T @ y
```

```python
f = lambda w: 1/ len(X) * np.linalg.norm(A @ w - y) ** 2
```

```python
g = lambda w: 1 / len(X) * 2 * A.T @ (A @ w - y)
```

```python
lr = 0.001
k = 0
w = np.array([[0], [0], [0]])
while k < 1_000:
    k += 1
    w = w - lr * g(w)
    e = np.linalg.norm(g(w))
    if k < 1e-3:
        print('*', k, w, e, f(w))
        break 
    print(k, w, e, f(w))
```
