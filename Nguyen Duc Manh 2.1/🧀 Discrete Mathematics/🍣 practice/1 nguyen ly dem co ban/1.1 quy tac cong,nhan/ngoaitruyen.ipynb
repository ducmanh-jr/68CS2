```python
# x1+x2+x3=0 x1< 15
def tim_nghiem():
    ket_qua = []
    for x1 in range(15):  # x1 có thể nhận giá trị từ 0 đến 14
        tong_con_lai = 15 - x1
        for x2 in range(tong_con_lai + 1):  # x2 có thể nhận giá trị từ 0 đến tong_con_lai
            x3 = tong_con_lai - x2  # x3 được xác định dựa trên x1 và x2
            ket_qua.append((x1, x2, x3))
    return ket_qua

# Tìm các nghiệm
nghiem = tim_nghiem()

# Đếm số lượng nghiệm
so_luong_nghiem = len(nghiem)
print(f"Số lượng nghiệm là: {so_luong_nghiem}")

#tìm nghiệm thứ 100 
nghiem100 = nghiem[99]
print(f"nghiệm thứ 100 {nghiem100}")

# In các nghiệm
for bo_nghiem in nghiem:
    print(bo_nghiem)
```

```python
# có 2 bi xanh3 bi đổ và 5 bi vàng làm sao để lấy ra 2 bi khác màu 
# Các viên bi
green_balls = ['G1', 'G2']  # 2 viên xanh
red_balls = ['R1', 'R2', 'R3']  # 3 viên đỏ
yellow_balls = ['Y1', 'Y2', 'Y3', 'Y4', 'Y5']  # 5 viên vàng

# Danh sách để lưu các cặp khác màu
pairs = []

# Lấy cặp xanh - đỏ
for g in green_balls:
    for r in red_balls:
        pairs.append((g, r))

# Lấy cặp xanh - vàng
for g in green_balls:
    for y in yellow_balls:
        pairs.append((g, y))

# Lấy cặp đỏ - vàng
for r in red_balls:
    for y in yellow_balls:
        pairs.append((r, y))

# In ra các cặp khác màu
print("Các cặp viên bi khác màu là:")
for pair in pairs:
    print(pair)


```
